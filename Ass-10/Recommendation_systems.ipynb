```python
len(b1['Book.Title'].unique())
```

pivoting the table

```python
b2=b1.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
b2
```

```python
b2.index=b1['User.ID'].unique()
b2
```

```python
b2.fillna(0,inplace=True)
```

Calculating Cosine Similarity between Users on array data

```python
from sklearn.metrics import pairwise_distances
user_sim=1-pairwise_distances(b2.values,metric='cosine')
user_sim
```

```python
us2=pd.DataFrame(user_sim)
us2
```

```python
us2.index=b1['User.ID'].unique()
us2.columns=b1['User.ID'].unique()
us2
```

Nullification

```python
np.fill_diagonal(user_sim,0)
us2
```

Similar user

```python
us2.idxmax(axis=1)
```
