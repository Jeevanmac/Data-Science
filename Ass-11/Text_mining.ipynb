```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
import re
import string
def clean_text(text):
  text=re.sub('@[A-Za-z0-9]+','',str(text))
  text=re.sub('#','',str(text))
  text=re.sub('RT[\s]+','',str(text))
  text=re.sub('\n\n','',str(text))
  text=text.lower()
  text=re.sub('https?:\/\/\S+','',str(text))
  text=text.translate(str.maketrans('', '', string.punctuation))

  return text
```

```python
elon['Text']=elon["Text"].apply(clean_text)
elon
```

```python
e2 = elon.Text
estring = ' '.join(e2)
estring
```

```python
from nltk.tokenize import word_tokenize
text_tokens=word_tokenize(estring)
print(text_tokens)
```

```python
with open('stop.txt', 'r') as stop:
  s1 = stop.read()
  s1=s1.split('\n')
```

```python
tweets=[word for word in text_tokens if word not in s1]
tweets
```

Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
textfit = TfidfVectorizer()
tf = textfit.fit_transform(tweets)
```

```python
feature_names=textfit.get_feature_names_out()
dense=tf.todense()
denselist=dense.tolist()
d=pd.DataFrame(denselist,columns=feature_names)
d
```

```python
list = ' '.join(d)
```

```python
wordcloud=WordCloud(background_color='black',width=1800,height=1400).generate(list)
plt.imshow(wordcloud)
```

```python
from textblob import TextBlob
def sentiment_analysis(ds):
    sentiment = TextBlob(ds["Text"]).sentiment
    return pd.Series([sentiment.subjectivity, sentiment.polarity])

elon[["subjectivity", "polarity"]] = elon.apply(sentiment_analysis, axis=1)
elon
```

```python
def analysis(score):
    if score < 0:
        return "Negative"
    elif score > 0:
        return "Positive"
    else:
        return "Neutral"

elon["analysis"] = elon["polarity"].apply(analysis)
elon
```
