* We are going to do 2-Tailed t-test
* significance = 5%
* alpha = 0.05 (cut-off)

```python
import scipy.stats as stats
import statsmodels.api as sm
statistic , p_value = stats.ttest_ind(ct['Unit A'],ct['Unit B'], alternative = 'two-sided')
print('p_value=',p_value)
```

**As p not equal to Significance.**
we fail to reject the Null Hypothesis

```python
alpha = 0.025
print('Significnace=%.3f, p=%.3f' % (alpha, p_value))
if p_value <= alpha:
    print('We reject Null Hypothesis')
else:
    print('We fail to reject Null hypothesis')
```

***As Null Hypothesis is rejected we can conclude that,there is no significant difference between the two samples***
