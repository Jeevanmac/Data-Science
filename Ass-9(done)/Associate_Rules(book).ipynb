```python
book = pd.read_csv('book.csv')
book
```

```python
book.info()
```

Association rules with 10% Support

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
fi = apriori(book,min_support=0.1,use_colnames=True)
fi
```

Association rules with 70% confidence

```python
ar = association_rules(fi,metric='lift',min_threshold=0.7)
ar
```

1. A leverage value of 0 indicates independence. Range will be [-1 1]
2. A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]


```python
ar.sort_values('lift',ascending=False)
```

Lift Ratio

```python
ar[ar.lift>1]
```

```python
plt.scatter(ar['support'],ar['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
