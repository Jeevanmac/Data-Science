# Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
sns.pairplot(tc2)
```

# Model Evaluation

```python
import statsmodels.formula.api as smf
model = smf.ols("Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight",data=tc2).fit()
```

```python
model.params
```

```python
model.summary()
```

Testing

```python
t_d = tc2.sample()
```

```python
t_d
```

```python
model.predict(t_d)
```

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

* c contain Cooks
* _ might represent other information that is being ignored or not explicitly used in this context.

```python
sns.set_style(style='darkgrid')
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(tc2)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
```

```python
from statsmodels.graphics.regressionplots import influence_plot
fig,ax=plt.subplots(figsize=(15,10))
fig=influence_plot(model,ax = ax)
```

* (80) plot don't have more influence on the regression model.
* Hence, it is neglected.

## Model building

```python
sp1 = sp.rename({'R&D Spend': 'RDS','Marketing Spend': 'Marketing'}, axis=1)
```

```python
model1 = smf.ols('Profit~RDS+Administration+Marketing',data=sp1).fit()
```

```python
model1.params
```

```python
model1.summary()
```

**Calculating R-squared for Individual Regression Models:**

```python
rsq_rds = smf.ols('RDS~Administration+Marketing',data=sp1).fit().rsquared
vif_rds = 1/(1-rsq_rds)
vif_rds

rsq_Administration = smf.ols('Administration~RDS+Marketing',data=sp1).fit().rsquared
vif_Administration = 1/(1-rsq_Administration)
vif_Administration


rsq_Marketing = smf.ols('Marketing~RDS+Administration',data=sp1).fit().rsquared
vif_Marketing = 1/(1-rsq_Marketing)
vif_Marketing
```

```python
dict = {'Variables':['RDS','Administration','Marketing'],'VIF':[vif_rds,vif_Administration,vif_Marketing]}
Vif_frame = pd.DataFrame(dict)
Vif_frame
```

* now we plot influence plots
* This is being done to spot any extra influence by a single data point on the model and remove that datapoint to improve its accuracy.


```python
model1_influence = model1.get_influence()
(c, _) = model1_influence.cooks_distance
```

```python
sns.set_style(style='darkgrid')
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(sp1)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
```

```python
fig,ax=plt.subplots(figsize=(15,10))
fig=influence_plot(model1,ax = ax)
```

* Here we see that the datapoint at index number 49 is having way too much influence on the model
* Hence we remove this entry and make a new model.
  

```python
sp_n=sp1.drop(sp1.index[[49]],axis=0).reset_index(drop=True)
sp_n
```

```python
New_model = smf.ols('Profit~RDS+Administration+Marketing',data=sp_n).fit()
```

```python
New_model.params,New_model.summary()
```

```python
new_test = sp_n.sample()
new_test
```

```python
New_model.predict(new_test)
```

```python
Tab={'Prepared_models':['model1','new_model'],'Rsquared':[model1.rsquared,New_model.rsquared]}
table=pd.DataFrame(Tab)
table
```
